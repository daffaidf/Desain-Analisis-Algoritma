Soal 1


```python
def hitung_angka_genap(arr):
    jumlah = 0
    for angka in arr:
        if angka % 2 == 0:
            jumlah += 1
    return jumlah

print(hitung_angka_genap([1, 2, 4, 7, 10]))
```

1. Pendekatan/Algoritma:
Iterasi setiap elemen array dan cek dengan modulo (%). Jika angka % 2 == 0, maka genap dan counter bertambah. Kompleksitas: O(n).

2. Edge Cases:
Array kosong → return 0
Angka 0 → termasuk genap
Angka negatif (-2, -4) → termasuk genap
Array 1 elemen → tetap berfungsi normal

soal 2

```python
def cari_nilai_terbesar(arr):
    if len(arr) == 0:
        return None

    terbesar = arr[0]
    for angka in arr:
        if angka > terbesar:
            terbesar = angka
    return terbesar

print(cari_nilai_terbesar([5, 2, 9, 1]))
```

1. Langkah-langkah:
Inisialisasi terbesar dengan elemen pertama, iterasi array, update terbesar jika menemukan angka yang lebih besar. Kompleksitas: O(n).

2. Array kosong:
Return None atau raise ValueError("Array tidak boleh kosong"). Pilihan tergantung kebutuhan aplikasi.

soal 3

```python
def balik_string(s):
    return s[::-1]

print(balik_string("halo"))
```

1. Logika:
Gunakan slicing s[::-1] yang berarti iterasi dari akhir ke awal dengan step -1. Atau bisa manual dengan loop dari indeks terakhir ke 0. Kompleksitas: O(n).


soal 4

```python
def jumlahkan_elemen(arr):
    total = 0
    for angka in arr:
        total += angka
    return total

print(jumlahkan_elemen([3, 1, 2]))
```

1. Cara kerja:
Inisialisasi total = 0. Iterasi array dan tambahkan setiap elemen ke total. Untuk [3,1,2]: total = 0+3=3, lalu 3+1=4, lalu 4+2=6. Return 6. Kompleksitas: O(n).

soal 5

```python
def cek_palindrome(s):
    s = s.lower()
    return s == s[::-1]

print(cek_palindrome("level"))
print(cek_palindrome("hello"))
```

1. Cara menentukan:
Normalisasi string ke lowercase, lalu bandingkan string asli dengan versi terbaliknya (s[::-1]). Jika sama = palindrome. Kompleksitas: O(n).

soal 6

```python
def hitung_vokal(s):
    vokal = "aiueoAIUEO"
    jumlah = 0
    for karakter in s:
        if karakter in vokal:
            jumlah += 1
    return jumlah

print(hitung_vokal("informatika"))
```

1. Cara memeriksa:
Definisikan string vokal "aiueoAIUEO". Iterasi setiap karakter input, cek apakah ada di string vokal dengan operator "in". Jika ya, tambah counter. Kompleksitas: O(n).

soal 7

```python
def cari_elemen(arr, target):
    return target in arr

print(cari_elemen([4, 7, 1, 9], 7))
print(cari_elemen([4, 7, 1, 9], 5))
```

1. Metode pencarian:
Linear Search: iterasi setiap elemen dan bandingkan dengan target. Jika ketemu return True, jika sampai akhir tidak ketemu return False. Kompleksitas: O(n). Jika array terurut, bisa pakai Binary Search O(log n).
