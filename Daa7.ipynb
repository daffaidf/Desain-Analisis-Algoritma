```python
def nilai_maksimal (deret_bilangan):
    nilai_terbesar = deret_bilangan[0]

    for nilai in deret_bilangan:
        if nilai > nilai_terbesar:
            nilai_terbesar = nilai

    return nilai_terbesar
```

```python
a = [3, 20,100, -35, 50]

print(a)
print('Nilai terbesar:', nilai_maksimal(a))
```

```python
def sequentialSearch(alist, item):
    pos = 0
    found = False

    while pos < len(alist) and not found:
        if alist[pos] == item:
            found = True
        else:
            pos = pos + 1

    return found

testlist = [1, 2, 32, 8, 17, 19, 42, 13, 0]
print(sequentialSearch(testlist, 3))
print(sequentialSearch(testlist, 2))
```

```python
from math import sqrt
from random import randint

titik = []
n = int(input('Masukkan jumlah titik yang ingin anda cari jaraknya: '))

# Membuat titik acak
for i in range(n):
    titik.append([randint(0, 100), randint(0, 100)])

print('Titik:\n', titik)

# Fungsi untuk menghitung jarak dan mencari jarak terdekat
def hitungjarak(lis):
    lisjarak = []
    for i in range(len(lis) - 1):
        for j in range(i + 1, len(lis)):
            jarak = sqrt((lis[i][0] - lis[j][0])**2 + (lis[i][1] - lis[j][1])**2)
            lisjarak.append(jarak)
            print('Titik:', lis[i], 'dan', lis[j], 'jaraknya adalah:', jarak)
    return min(lisjarak)

print('Jarak terdekat:\n', hitungjarak(titik))
```

```python
def string_match(string, sub_str):
    # Brute force string matching
    for i in range(len(string) - len(sub_str) + 1):
        match = True
        for j in range(len(sub_str)):
            if string[i + j] != sub_str[j]:
                match = False
                break
        if match:
            return i
    return -1

if __name__ == "__main__":
    print(string_match("adbcbdc", "bdc"))

```

```python
import random
from itertools import permutations

# Semua kemungkinan urutan kota (tur)
alltours = permutations

def distance_tour(aTour):
    """Menghitung total jarak dari satu tur."""
    return sum(distance_points(aTour[i - 1], aTour[i]) for i in range(len(aTour)))

def distance_points(first, second):
    """Menghitung jarak Euclidean antara dua titik (kompleks)."""
    return abs(first - second)

def aCity(x, y):
    """Membuat representasi kota sebagai bilangan kompleks."""
    return complex(x, y)

def generate_cities(number_of_cities):
    """Menghasilkan sekumpulan kota acak dalam bidang 2D."""
    seed = 111
    width = 500
    height = 300
    random.seed((number_of_cities,  + seed))
    return frozenset(
        aCity(random.randint(1, width), random.randint(1, height))
        for c in range(number_of_cities)
    )
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def visualize_tour(tour, style='bo-'):
    if len(tour) > 1000:
        plt.figure(figsize=(15, 10))
    start = tour[0:1]
    visualize_segment(tour + start, style)
    visualize_segment(start, 'rD')

def visualize_segment(segment, style='bo-'):
    plt.plot([X(c) for c in segment], [Y(c) for c in segment], style, clip_on=False)
    plt.axis('scaled')
    plt.axis('off')

def X(city):
    """Mengambil koordinat X (sumbu real)."""
    return city.real

def Y(city):
    """Mengambil koordinat Y (sumbu imajiner)."""
    return city.imag
```

```python
from time import process_time
from collections import Counter

def tsp(algorithm, cities):
    """Menjalankan algoritma TSP dan menampilkan hasil visualisasi serta waktu eksekusi."""
    t0 = process_time()
    tour = algorithm(cities)
    t1 = process_time()

    # Pastikan semua kota dikunjungi tepat satu kali
    assert Counter(tour) == Counter(cities)

    # Visualisasi hasil tur
    visualize_tour(tour)

    # Cetak hasil
    print("{}: {} cities => tour length {:.0f} (in {:.3f} sec)".format(
        name(algorithm), len(tour), distance_tour(tour), t1 - t0
    ))

def name(algorithm):
    """Mengambil nama algoritma tanpa suffix '_tsp'."""
    return algorithm.__name__.replace('_tsp', '')
```
