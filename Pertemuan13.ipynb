```python
graph = {
    'Amin': ['Wasim', 'Nick', 'Mike'],
    'Wasim': ['Imran', 'Amin'],
    'Imran': ['Wasim', 'Faras'],
    'Faras': ['Imran'],
    'Mike': ['Amin'],
    'Nick': ['Amin']
}

def bfs(graph, start_node):
    visited = []
    queue = [start_node]

    while queue:
        node = queue.pop(0)

        if node not in visited:
            visited.append(node)

            neighbours = graph.get(node, [])

            for neighbour in neighbours:
                if neighbour not in visited:
                    queue.append(neighbour)

    return visited

def dfs(graph, start_node, visited=None):
    if visited is None:
        visited = set()

    if not hasattr(dfs, 'path'):
        dfs.path = []

    if start_node not in visited:
        dfs.path = []

    def _dfs_recursive(node):
        visited.add(node)
        dfs.path.append(node)

        neighbours = graph.get(node, [])

        for neighbour in neighbours:
            if neighbour not in visited:
                _dfs_recursive(neighbour)

    _dfs_recursive(start_node)

    result_path = dfs.path
    del dfs.path
    return result_path

print("--- Data Graf ---")
print(graph)
print("\n")

print("====================================")
print("HASIL BREADTH-FIRST SEARCH (BFS)")
print("====================================")

result_amin_bfs = bfs(graph, 'Amin')
print(f"BFS dari 'Amin': {result_amin_bfs}")

result_wasim_bfs = bfs(graph, 'Wasim')
print(f"BFS dari 'Wasim': {result_wasim_bfs}")

result_faras_bfs = bfs(graph, 'Faras')
print(f"BFS dari 'Faras': {result_faras_bfs}")


print("\n")


print("====================================")
print("HASIL DEPTH-FIRST SEARCH (DFS)")
print("====================================")

result_amin_dfs = dfs(graph, 'Amin')
print(f"DFS dari 'Amin': {result_amin_dfs}")

result_wasim_dfs = dfs(graph, 'Wasim')
print(f"DFS dari 'Wasim': {result_wasim_dfs}")

result_faras_dfs = dfs(graph, 'Faras')
print(f"DFS dari 'Faras': {result_faras_dfs}")
```
